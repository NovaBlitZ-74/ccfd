```python
#Input Split
X = df.drop(columns=['Class'], axis=1)
Y = df['Class']
```

```python
#Standard Scaling
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
X_scaler= sc.fit_transform(X)
```

```python
#train test split
from sklearn.model_selection import train_test_split
X_train ,X_test, Y_train, Y_test = train_test_split(X_scaler, Y, test_size=0.25, random_state=42, stratify=Y)
```

## Implementing models without Balancing

```python
#logistic regression
from sklearn.metrics import classification_report,f1_score
from sklearn.linear_model import LogisticRegression
logreg= LogisticRegression(random_state=42)
logreg.fit(X_train, Y_train)
Y_pred_lr = logreg.predict(X_test)
print(classification_report(Y_test, Y_pred_lr))
print(f1_score(Y_test, Y_pred_lr))
```

```python
import tensorflow
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from sklearn.utils import class_weight
from sklearn.utils import compute_class_weight
from scikeras.wrappers import KerasClassifier
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.metrics import roc_auc_score, roc_curve, average_precision_score
from sklearn.model_selection import cross_val_score, cross_validate
from sklearn.model_selection import GridSearchCV

train_classes = Y_train
class_weights = compute_class_weight(
                                        class_weight = "balanced",
                                        classes = np.unique(train_classes),
                                        y = train_classes                                                    
                                    )
class_weights = dict(zip(np.unique(train_classes), class_weights)),
print("class_weights : ", class_weights)
```

```python
print(len(X.columns))
n_inputs=30
```

```python
#Neural network model
model= Sequential()
model.add(Dense(64, input_dim=n_inputs, activation='relu', kernel_initializer='he_uniform'))
model.add(Dense(32,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='Adam')
class_weights = {0: 0.5, 1: 289}
model.fit(X_train, Y_train, validation_split = 0.1,epochs = 20)
```

```python
from sklearn.metrics import roc_auc_score
Y_pred_nn= model.predict(X_test)

print(roc_auc_score(Y_test,Y_pred_nn))
```

```python
#Decision Tree model
from sklearn.tree import DecisionTreeClassifier
dectree = DecisionTreeClassifier()
dectree.fit(X_train, Y_train)
Y_pred_dt= dectree.predict(X_test)
print(classification_report(Y_test, Y_pred_dt))
print(f1_score(Y_test, Y_pred_dt))
```

```python
#Random Forest model
from sklearn.ensemble import RandomForestClassifier
randfor = RandomForestClassifier()
randfor.fit(X_train, Y_train)
Y_pred_rf= randfor.predict(X_test)
print(classification_report(Y_test, Y_pred_rf))
print(f1_score(Y_test, Y_pred_rf))
```

```python
#Gradient Boosting model
from sklearn.ensemble import GradientBoostingClassifier
gradbst = GradientBoostingClassifier()
gradbst.fit(X_train, Y_train)
Y_pred_gb= gradbst.predict(X_test)
print(classification_report(Y_test, Y_pred_gb))
print(f1_score(Y_test, Y_pred_gb))
```

```python
# XGBOOST model
from xgboost import XGBClassifier
xgb = XGBClassifier()
xgb.fit(X_train, Y_train)
Y_pred_xgb = xgb.predict(X_test)
print(classification_report(Y_test, Y_pred_xgb))
print(f1_score(Y_test, Y_pred_xgb))
```

```python
#Lightgbm model
from lightgbm import LGBMClassifier
lgbm = LGBMClassifier()
lgbm.fit(X_train, Y_train)
Y_pred_lgbm = lgbm.predict(X_test)
print(classification_report(Y_test, Y_pred_lgbm))
print(f1_score(Y_test, Y_pred_lgbm))
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Example: Compute and store metrics for each model
logreg_metrics = {
    'accuracy': accuracy_score(Y_test, Y_pred_lr),
    'precision': precision_score(Y_test, Y_pred_lr),
    'recall': recall_score(Y_test, Y_pred_lr),
    'f1-score': f1_score(Y_test, Y_pred_lr)
}

randomforest_metrics = {
    'accuracy': accuracy_score(Y_test, Y_pred_rf),
    'precision': precision_score(Y_test, Y_pred_rf),
    'recall': recall_score(Y_test, Y_pred_rf),
    'f1-score': f1_score(Y_test, Y_pred_rf)
}

lightgbm_metrics = {
    'accuracy': accuracy_score(Y_test, Y_pred_lgbm),
    'precision': precision_score(Y_test, Y_pred_lgbm),
    'recall': recall_score(Y_test, Y_pred_lgbm),
    'f1-score': f1_score(Y_test, Y_pred_lgbm)
}

# Store all metrics in a single dictionary
performance_metrics = {
    'Logistic Regression': logreg_metrics,
    'Random Forest': randomforest_metrics,
    'LightGBM': lightgbm_metrics
}

# Extract the metric names and model names
metrics = list(performance_metrics['Logistic Regression'].keys())
models = list(performance_metrics.keys())

# Prepare the data for plotting
data = np.array([[performance_metrics[model][metric] for model in models] for metric in metrics])

# Plotting
fig, ax = plt.subplots(figsize=(10, 6))
bar_width = 0.2
index = np.arange(len(metrics))

for i, model in enumerate(models):
    ax.bar(index + i * bar_width, data[:, i], bar_width, label=model)

ax.set_xlabel('Metrics')
ax.set_ylabel('Scores')
ax.set_title('Performance Metrics Comparison')
ax.set_xticks(index + bar_width)
ax.set_xticklabels(metrics)
ax.legend()

plt.show()
```
